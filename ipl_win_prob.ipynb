```python
delivery_df['Score'] = delivery_df[['match_id', 'ball_score']].groupby('match_id').cumsum()['ball_score']
```

```python
#new column target_left
delivery_df['target_left'] = (delivery_df['total_runs']+1) - delivery_df['Score']
```

```python
delivery_df
```

```python
#balls remaininig
delivery_df['Remaining Balls'] = 120 - ((delivery_df['over'] - 1)*6 + delivery_df['ball'])
```

```python
delivery_df
```

```python
#how many wicktes are left out
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna('0')
```

```python
delivery_df
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x: x if x == '0' else 1).astype(int)
```

```python
delivery_df[['match_id', 'player_dismissed']]
```

```python
delivery_df['Wickets'] = delivery_df[['match_id', 'player_dismissed']].groupby('match_id').cumsum()['player_dismissed'].values
```

```python
delivery_df
```

```python
delivery_df['Wickets Left'] = 10 - delivery_df['Wickets']
```

```python
delivery_df
```

Current Run Rate  

$$Current \ run \ rate = \frac{Runs \ scored} {Number \ of \ Overs \ completed}$$

```python
#current run rate
delivery_df['CRR'] = (delivery_df['Score']*6 / (120 - delivery_df['Remaining Balls']))
```

Required Run Rate  

$$Required \ run \ rate = \frac{Remaining \ scored \ or \ Target} {Number \ of \ Overs \ left}$$

```python
#required run rate
delivery_df['RRR'] = (delivery_df['target_left']*6 / delivery_df['Remaining Balls'])
```

```python
delivery_df
```

```python
#creating a new column result where 1 means won and 0 menas lost
def result(row):
    if row['batting_team'] == row['winner']:
        return 1
    else:
        return 0
```

```python
delivery_df['result'] = delivery_df.apply(result, axis=1)
```

```python
model_df = delivery_df[['batting_team', 'bowling_team', 'city', 'Score', 'Wickets Left','Remaining Balls','target_left','CRR', 'RRR', 'result']]
```

```python
model_df.isnull().sum()
```

```python
model_df.dropna(inplace=True)
```

Since we are getting -inf and +inf in the rrr column that means as per the formula the number of balls remaning are 0. hence we will consider only those records where the remaining balss are not 0

```python
model_df = model_df[model_df['Remaining Balls'] != 0]
```

```python
model_df.describe()
```

```python
#shuffling the data so that the model does not learn the order of the data or become biased
model_df = model_df.sample(model_df.shape[0])
```

```python
#splitting the data into features and target
X = model_df.drop('result', axis=1)
y = model_df['result']
```

```python
X
```

```python
y
```

```python
#spiltting the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
#label encoding or one hot encoding (since no hirearchial data we shall use one hot encoding)
#transform non-numerical data into numerical data
from sklearn.compose import ColumnTransformer 
from sklearn.preprocessing import OneHotEncoder
trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse_output = False, drop = 'first'),['batting_team', 'bowling_team', 'city'])
], remainder = 'passthrough')
```

```python
trf
```

```python
#buikding the model
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps = [
    ('step1',trf),
    ('step2', LogisticRegression(solver = 'liblinear'))
])
```

```python
pipe.fit(X_train, y_train)
```

```python
Y_pred = pipe.predict(X_test)
```

```python
Y_pred
```

```python
#accuracy score
from sklearn.metrics import accuracy_score
accuracy_score(y_test, Y_pred)
```
